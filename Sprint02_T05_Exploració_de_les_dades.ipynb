```python
# Resumeix estadísticament les columnes d’interès
print(df02.info())
```

```python
# Troba quantes dades faltants hi ha per columna
print(df02.isnull().sum())
```

```python
# Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)
df03=df02.assign(VelocitatMitjanaVol=df02["Distance"]/df02["AirTime"])
#df03=df02.assign(ArribaTard="Sí" if ["ArrDelay"]>0)
df03["ArribaTard"] = df03["ArrDelay"].apply(lambda x: "Yes" if x > 0 else "No")

print(df03)
print(df03.info())
```

```python
# Taula de les aerolínies amb més endarreriments acumulats
df04=df03.groupby("UniqueCarrier")[["ArrDelay"]].sum()
df04.info()
df04.sort_values(by="ArrDelay", ascending=False)
```

```python
# Quins són els vols més llargs? I els més endarrerits?
df05=df.iloc[:,np.r_[9, 14]]
filtre = df05["AirTime"] > 1
df06 = df05[filtre]
print(df06.sort_values(by="AirTime", ascending=False))
print("El vol més llarg és: " + str(df06.max()))
print("El vol més curt és: " + str(df06.min()))
```
